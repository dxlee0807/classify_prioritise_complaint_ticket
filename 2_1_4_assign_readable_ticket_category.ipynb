```python
import pandas as pd

df = pd.read_csv('lda_topic.csv')
df
```

# Assign readable Ticket Category

```python
#Create the dictionary of Topic IDs and readable Ticket Category
ticket_categories = {
    0:'Communication and Interactions',
    1:'Loans and Mortgages',
    2:'Consumer Financial Issues',
    3:'Credit Cards',
    4:'Banking and Financial Accounts',
    5:'Fraud Detection and Mitigation',
    6:'Credit and Credit Reports',
    7:'Payments and Billing',
    8:'Dispute Resolution',
}
df['category'] = df['topic_id'].map(ticket_categories)
```

# Export for deployment

```python
df.drop(columns='topic_name',inplace=True)
df.to_csv('ticket_category.csv',index=False)
```
