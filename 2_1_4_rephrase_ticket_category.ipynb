# Rephrase Ticket Category

```python
#Create the dictionary of Topic IDs and readable Ticket Category
ticket_categories = {
    0:'Credit Card Payment Management',
    1:'Fraudulent Transactions and Claims',
    2:'Credit Information Requests and Complaints',
    3:'Banking Security and Fraud Prevention',
    4:'Payment Fees and Interest Rates',
    5:'Credit Card Disputes and Refunds',
    6:'Bank Account Transactions',
    7:'Home Loan and Mortgage Management',
    8:'Communication and Inquiry',
}
df['category'] = df['topic_id'].map(ticket_categories)
```
